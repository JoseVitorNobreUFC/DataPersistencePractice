```python
import filetype
from bs4 import BeautifulSoup
from PyPDF2 import PdfReader
from PIL import Image
import pytesseract
import requests

# 3. Implementação Completa de um Extrator de Dados Estruturados e Não Estruturados

def detectar_tipo_arquivo(caminho):
    tipo = filetype.guess(caminho)
    if tipo:
        return tipo.mime
    if caminho.lower().endswith(('.html', '.htm')):
        return 'text/html'
    return 'desconhecido'

def extrair_html(caminho):
    with open(caminho, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f.read(), 'html.parser')
        return soup.get_text(separator='\n', strip=True)

def extrair_pdf(caminho):
    texto = []
    with open(caminho, 'rb') as f:
        leitor = PdfReader(f)
        for pagina in leitor.pages:
            texto.append(pagina.extract_text())
    return '\n'.join(texto)

def extrair_imagem(caminho):
    imagem = Image.open(caminho)
    return pytesseract.image_to_string(imagem)

def processar_arquivo(caminho):
    mime_type = detectar_tipo_arquivo(caminho)

    print(f"Processando: {caminho} | Tipo: {mime_type}")
    print("-" * 50)

    try:
        if 'html' in mime_type:
            resultado = extrair_html(caminho)
        elif 'pdf' in mime_type:
            resultado = extrair_pdf(caminho)
        elif 'image' in mime_type:
            resultado = extrair_imagem(caminho)
        else:
            return "Tipo de arquivo não suportado"

        return resultado.strip()

    except Exception as e:
        return f"Erro na extração: {str(e)}"


arquivos = [
    'pagina_web.html',
    'documento.pdf',
    'produto.png'
]

for arquivo in arquivos:
    resultado = processar_arquivo(arquivo)
    print(resultado)
    print("\n" + "="*50 + "\n")
```
